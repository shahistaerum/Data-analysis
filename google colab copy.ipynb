```python
df = df.dropna(subset=['Make'])
```

```python
df.nunique()
```

```python
plt.figure(figsize=(12, 6))
df['Make'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Electric Vehicle Manufacturers')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Model Year'], bins=20, kde=False)
plt.title('Distribution of EV Model Years')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Make', order=df['Make'].value_counts().index[:10], hue='Electric Vehicle Type')
plt.title('Top 10 Makes by Electric Vehicle Type')
plt.show()

```

```python
cat_columns = ['Make', 'Model', 'Electric Vehicle Type', 'Vehicle Location', 'County', 'City']
for col in cat_columns:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts().head(10))
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()
df[num_cols].hist(figsize=(15, 10), bins=20)
plt.suptitle('Histograms of Numerical Columns', fontsize=16)
plt.show()
```

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('cleaned_sorted_data.xlsx')

# 1 Sort columns alphabetically by their names (titles)
df = df[sorted(df.columns)]

# 2 Sort rows by the first column (after sorting column titles)
first_column = df.columns[0]
df_sorted = df.sort_values(by=first_column)

# 3 Export to Excel
df_sorted.to_excel('cleaned_sorted_final.xlsx', index=False)

# 4 Download the new file
from google.colab import files
files.download('cleaned_sorted_final.xlsx')
```
